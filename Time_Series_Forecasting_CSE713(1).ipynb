#**Time Series Forecasting with statsmodels**

```python
import pandas as pd

df = pd.read_csv("dc.csv")
df = df.rename(columns = {"Unnamed: 0": "Time"})
df["Time"] = pd.to_datetime(df["Time"])
df = df.iloc[::-1].set_index("Time")
df.head()
```

```python
import matplotlib.pyplot as plt
plt.title("CloseUSD against Time")
plt.xlabel("Time")
plt.ylabel("CloseUSD")
plt.plot(df['close_USD'])
plt.show()
```

```python
# Split the data
train = df.iloc[:-200]
test = df.iloc[-200:]
```

```python
from statsmodels.tsa.arima.model import ARIMA

#sample parameters
model = ARIMA(train.close_USD, order=(2, 1, 0))
results = model.fit()

# Make predictions for the test set
forecast = results.forecast(steps=200)
forecast
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

test_close_USD = test["close_USD"]

#mean absolute error
mae = mean_absolute_error(test_close_USD, forecast)

#root mean square error
mse = mean_squared_error(test_close_USD, forecast)
rmse = np.sqrt(mse)

#mean absolute percentage error
mape = (forecast - test_close_USD).abs().div(test_close_USD).mean()

print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAPE: {mape:.2f}%")
```

```python
plt.plot(train.index, train["close_USD"], label='Train')
plt.plot(test.index, test["close_USD"], label='Test')
plt.plot(forecast.index, forecast, label='Forecast')
plt.legend()
plt.show()
```

```python
# df_p = df.reset_index()[["Time", "close_USD"]].rename(
#     columns={"Time": "ds", "close_USD": "y"}
# )

df_p = df.reset_index()
df_p = df_p[["Time", "close_USD"]]
df_p = df_p.rename(columns={"Time": "ds", "close_USD": "y"})
df_p.head()
```

```python
import pandas as pd
from prophet import Prophet

model = Prophet()

# Fit the model
model.fit(df_p)

# create date to predict
future_dates = model.make_future_dataframe(periods=365)

# Make predictions
predictions = model.predict(future_dates)

predictions.head()
```

```python
model.plot(predictions)
```

```python
model.plot_components(predictions)
```

```python
from prophet.diagnostics import cross_validation, performance_metrics

# Perform cross-validation with initial 365 days for the first training data and the cut-off for every 180 days.

df_cv = cross_validation(model, initial='365 days', period='180 days', horizon = '365 days')

# Calculate evaluation metrics
res = performance_metrics(df_cv)

res
```

```python
from prophet.plot import plot_cross_validation_metric
#choose between 'mse', 'rmse', 'mae', 'mape', 'coverage'

plot_cross_validation_metric(df_cv, metric= 'mape')
```
